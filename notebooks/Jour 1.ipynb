Consultez cet exemple qui contient une liste de pays clients

````python
# Création d'une variable contenant une liste ordonnée de pays
pays_clients = ["France", "Allemagne", "Canada", "Japon"]

# Affichage de différents éléemnts de la liste
print("Liste de pays : "+str(pays_clients))

# Affichage sans d'un élement de la liste de type chaîne de caractère pas besoin de conversion
print("Premier pays de la liste : "+ pays_clients[0])
print("Second pays de la liste : "+ pays_clients[1])
print("Dernier pays de la liste : "+ pays_clients[-1])

# Affichage de plusieurs élements de la liste, une conversion est néssaire pour l'afficher
print("Du début jusqu'au second : "+ str(pays_clients[:2]))
print("Du second jusqu'au troisième : "+ str(pays_clients[1:3]))


# la fonction len() affiche la taille d'une liste
# La fonction str() permet de convertir n'importe quel type de donnée en chaîne de caractère
print("Taille de la liste : " + str(len(pays_clients)))

# Ajout d'un élément dans la liste
pays_clients.append("Belgique")
# Affichage de la nouvelle liste de pays modifiée
print(pays_clients)

# Suppression d'un élément dans la liste
pays_clients.remove("Canada")
# Affichage de la nouvelle liste de pays modifiée
print(pays_clients)


#Faire la somme d'une liste :
somme = sum( [120, 340, 275, 410])
print("Somme totale : " + str(somme) )

````

## 🟣 Les fonctions (≈ 30 min)

Regardez cet exemple d'utilisation de la fonction total_commande:


````python

ventes = {
    "BioMarché": {"Quantité" : 120, "Prix Unitaire" : 6.71},
    "GreenFood": {"Quantité" : 340, "Prix Unitaire" : 6.23},
    "NutriPlus": {"Quantité" : 275, "Prix Unitaire" : 6.54},
    "EcoStore": {"Quantité" : 410, "Prix Unitaire" : 6.20},
    "VeggieWorld": {"Quantité" : 95, "Prix Unitaire" : 6.68}
}

def total_commande(quantite, prix_unitaire):
    resultat = quantite*prix_unitaire
    return resultat


montant_total = 0
for client in ventes :
    quantite = ventes[client]["Quantité"]
    prix_unitaire = ventes[client]["Prix Unitaire"]
    montant_total = montant_total + total_commande(quantite,prix_unitaire)
    # peut s'écrire aussi : montant_total += total_commande(quantite,prix_unitaire)
    
print("Montant des commandes cumulées " + str(montant_total) )

````
